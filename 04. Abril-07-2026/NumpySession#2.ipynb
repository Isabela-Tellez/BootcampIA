### **Array Broadcasting**

**VECTOR 1D
MATRIZ 2 o más Dimensiones**

Broadcasting es una característica en Numpy que permite ejecutar operaciones con arrays de diferentes formas y tamaños sin recorrer explicitamente elementos. Broadcasting automáticamente expande pequeños arrays para emparejar/encajar redimensionarlos al tamaño del grande.

**Reglas**
- Si el array es diferente en dimensiones, el pequeño es


```python
a = np.array([1, 2, 3])
b = 2

#SUMA
#Suma a cada elemento del array 2
result = a + b
print(result)
```

```python
#MATRIZ
#Le suma la dimension del arr a cada dimension de la matriz
arr = np.array([1, 2, 3])
matriz =np.array([
    [1, 2, 3],
    [4, 5, 6]])

result = arr + matriz
print(result)
```

## **Sistemas Lineales**: El Misterio del Ticket Borrado

Imaginemos que el dueño de la cafetería ha perdido la lista de precios y solo cuenta con los siguientes datos:

- Lunes: vendió 2 Cafés y 1 Croissant para un total de 5€
- Martes: vendió 1 Café y 1 Croissant para un total de 3€

¿Cuánto vale el Café?
¿Cuánto el Croissant?

```python
#Código Sistemas Lineales

#A Array
A = np.array([
    [2,1],#lunes
    [1,1],#martes
])

#B De lo que se paga
B = np.array([5, 3])

result = np.linalg.solve(A, B)
print(f"Precio Café: {result[0]}")
print(f"Precio Croissant: {result[1]}")
```
