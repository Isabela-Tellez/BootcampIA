```python
# --- DEMO DE CÓDIGO (GESTIÓN DE PEDIDOS) ---

# 1. Definición de los orígenes de datos según su estado
# Pedidos que ya han salido del almacén (Ventas confirmadas)
pedidos_enviados = [
    {"id": 101, "producto": "Laptop", "monto": 800},
    {"id": 102, "producto": "Mouse", "monto": 25},
    {"id": 103, "producto": "Teclado", "monto": 60}
]

# Pedidos que están en el sistema pero aún no se han procesado
pedidos_pendientes = [
    {"id": 104, "producto": "Monitor", "monto": 300},
    {"id": 105, "producto": "Webcam", "monto": 50}
]

# Pedidos que fueron cancelados o devueltos (Dinero que sale o nunca entró)
pedidos_devueltos = [
    {"id": 106, "producto": "Auriculares", "monto": 40},
    {"id": 107, "producto": "Cable USB", "monto": 10}
]

# 2. Procesamiento de los datos (Cálculo de sumas por categoría)
total_real = sum(p["monto"] for p in pedidos_enviados)
total_pendiente = sum(p["monto"] for p in pedidos_pendientes)
total_devuelto = sum(p["monto"] for p in pedidos_devueltos)

# 3. Demostración del error de análisis por no entender el ORIGEN
# ❌ ERROR: Tratar todos los orígenes como si fueran "Ventas Exitosas"
total_incorrecto = total_real + total_pendiente + total_devuelto

# ✅ ACIERTO: Filtrar y usar solo los orígenes que representan ingresos reales
total_correcto = total_real + total_pendiente

# 4. Presentación de resultados
print(f"Ventas reales enviadas: ${total_real}")
print(f"Ventas pendientes: ${total_pendiente}")
print(f"Ventas devueltas: ${total_devuelto}")

print(f"\n❌ Total INCORRECTO (todo mezclado): ${total_incorrecto}")
print(f"✅ Total CORRECTO (sin devueltos): ${total_correcto}")

# Conclusión visual del impacto del error
print(f"\nDiferencia: ${total_incorrecto - total_correcto} (dinero que NO ingresó)")
```

```python
import pandas as pd

# --- SIMULACIÓN DE ORIGEN DE DATOS (MARKETING) ---

# Creamos el diccionario que representa la fuente original: Un formulario web.
# Observad que en el ORIGEN ya existen correos repetidos (ana@gmail.com).
data_marketing = {
    'email': ['ana@gmail.com', 'pedro@hotmail.com', 'ana@gmail.com', 'luis@yahoo.com', 'ana@gmail.com'],
    'fecha_registro': ['2024-03-01', '2024-03-01', '2024-03-01', '2024-03-02', '2024-03-02']
}

# Convertimos ese origen de datos en un DataFrame de Pandas para poder analizarlo.
df_marketing = pd.DataFrame(data_marketing)

# Mostramos los datos tal cual llegan de la fuente para que la audiencia vea el error.
print("--- DATOS RECIBIDOS DEL FORMULARIO ---")
print(df_marketing)

# --- ANÁLISIS DEL ORIGEN ---

# Calculamos cuántas filas han entrado en total (Registros brutos).
total_registros = len(df_marketing)

# Usamos la función .nunique() para contar cuántos correos diferentes (únicos) hay realmente.
usuarios_unicos = df_marketing['email'].nunique()

# Imprimimos los resultados para demostrar el impacto del problema de origen.
print(f"\nAnálisis rápido:")
print(f"-> Registros totales en el origen: {total_registros}")
print(f"-> Clientes reales (únicos): {usuarios_unicos}")

# Calculamos y mostramos el porcentaje de "ruido" o duplicidad que viene de la fuente.
print(f"-> ¡Cuidado! El origen nos está duplicando un {((total_registros-usuarios_unicos)/total_registros)*100}% de los datos.")
```

```python
# primero se comprueba si los datos vienen bien formados y luego se calcula algo con ellos.
datos = [
    {"nombre": "Ana", "nota": 8},
    {"nombre": "Luis", "nota": 7},
    {"nombre": "Ana", "nota": 8}  # dato duplicado
]
nombres = [d["nombre"] for d in datos]
if len(nombres) != len(set(nombres)):
    print("Aviso: hay datos duplicados, revisa el origen.")
else:
    media = sum(d["nota"] for d in datos) / len(datos)
    print("Nota media:", media)

# Este código muestra que no basta con calcular la media: antes hay que mirar si los datos vienen limpios y si el origen es fiable.
# Aquí el problema es que Ana aparece dos veces, así que el resultado puede parecer válido aunque la entrada ya esté mal.
```
