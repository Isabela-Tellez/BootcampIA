```python
# Importamos la librería matplotlib.pyplot y la llamamos plt
# Esta librería sirve para crear gráficos
import matplotlib.pyplot as plt



# -----------------------------
# 1. GRÁFICO DE BARRAS
# -----------------------------
# Este gráfico sirve para comparar cantidades entre distintas categorías

# Lista con los nombres de los productos
productos = ['Portátiles', 'Ratones', 'Teclados', 'Monitores']

# Lista con las unidades vendidas de cada producto
ventas = [15, 40, 25, 10]

# Creamos una figura nueva y le damos un tamaño de 8x5
plt.figure(figsize=(8,5))

# Creamos el gráfico de barras
# En el eje X irán los productos
# En el eje Y irán las ventas
plt.bar(productos, ventas)

# Añadimos un título al gráfico
plt.title('Ventas por producto')

# Ponemos nombre al eje horizontal
plt.xlabel('Productos')

# Ponemos nombre al eje vertical
plt.ylabel('Unidades vendidas')

# Mostramos el gráfico en pantalla
plt.show()
```

```python
# Importamos la librería matplotlib.pyplot y la llamamos plt
# Esta librería sirve para crear gráficos
import matplotlib.pyplot as plt

# 2. GRÁFICO DE LÍNEAS
# -----------------------------
# Este gráfico sirve para mostrar cómo cambian los datos con el tiempo

# Lista con los días de la semana
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']

# Lista con el número de visitas de cada día
visitas = [120, 150, 180, 170, 210, 260, 230]

# Creamos una figura nueva
plt.figure(figsize=(8,5))

# Creamos el gráfico de líneas
# marker='o' hace que se vean los puntos sobre la línea
plt.plot(dias, visitas, marker='o')

# Añadimos un título
plt.title('Visitas a la web durante una semana')

# Nombre del eje X
plt.xlabel('Días')

# Nombre del eje Y
plt.ylabel('Número de visitas')

# Mostramos el gráfico
plt.show()
```

```python
# Importamos la librería matplotlib.pyplot y la llamamos plt
# Esta librería sirve para crear gráficos
import matplotlib.pyplot as plt
# -----------------------------
# 3. HISTOGRAMA
# -----------------------------
# Este gráfico sirve para ver cómo se distribuyen datos numéricos

# Lista con las notas de varios alumnos
notas = [4, 5, 6, 7, 8, 6, 5, 7, 9, 10, 4, 6, 7, 8, 5, 6, 9, 7, 8, 6]

# Creamos una figura nueva
plt.figure(figsize=(8,5))

# Creamos el histograma
# bins=7 divide los datos en 7 intervalos o grupos
plt.hist(notas, bins=7)

# Añadimos un título
plt.title('Distribución de notas')

# Nombre del eje X
plt.xlabel('Nota')

# Nombre del eje Y
plt.ylabel('Frecuencia')

# Mostramos el gráfico
plt.show()
```

```python
# Importamos la librería matplotlib.pyplot y la llamamos plt
# Esta librería sirve para crear gráficos
import matplotlib.pyplot as plt
#----------------------------
# 4. GRÁFICO DE DISPERSIÓN
# -----------------------------
# Este gráfico sirve para ver si existe relación entre dos variables

# Lista con las horas que han estudiado varios alumnos
horas_estudio = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]

# Lista con la nota obtenida por esos alumnos
nota_examen = [4, 5, 5, 6, 6, 7, 7, 8, 9, 9]

# Creamos una figura nueva
plt.figure(figsize=(8,5))

# Creamos el gráfico de dispersión
# Cada punto representa una pareja de valores:
# horas de estudio y nota obtenida
plt.scatter(horas_estudio, nota_examen)

# Añadimos un título
plt.title('Relación entre horas de estudio y nota')

# Nombre del eje X
plt.xlabel('Horas de estudio')

# Nombre del eje Y
plt.ylabel('Nota del examen')

# Mostramos el gráfico
plt.show()
```
