# **Introducción a la librería Pandas**

[Pandas](https://pandas.pydata.org/about/index.html) es una librería que proporciona estructuras de datos y herramientas de análisis de datos de alto rendimiento y fáciles de usar.
* La estructura de datos principal es el DataFrame, que puede considerarse como una tabla 2D en memoria (como una hoja de cálculo, con nombres de columna y etiquetas de fila).
* Muchas funciones disponibles en Excel están disponibles mediante programación, como crear tablas dinámicas, calcular columnas basadas en otras columnas, trazar gráficos, etc.
* Proporciona un alto rendimiento para manipular (unir, dividir, modificar…) grandes conjuntos de datos

[Guía Usuario (Documentación Oficial)](https://pandas.pydata.org/docs/user_guide/index.html#user-guide)

```python
import pandas as pd
print('✅ Pandas', pd.__version__, '— listo!')
```

## Estructuras de datos en Pandas

La librería Pandas, de manera genérica, contiene las siguientes estructuras de datos:
* **Series**: Array de una dimensión
* **DataFrame**: Se corresponde con una tabla de 2 dimensiones
* **Panel**: Similar a un diccionario de DataFrames

De NumPy a Pandas

Ya conocen los **arrays de NumPy**: una lista de números con superpoderes matemáticos.

Pandas añade una capa encima:

| NumPy | Pandas |
|-------|--------|
| Array 1D | **Series** (array + etiquetas) |
| Array 2D | **DataFrame** (tabla completa) |
| Solo números | Números, texto, fechas, booleanos… |
| Sin nombres de columna | Columnas con nombre |

🧠 **Analogía:** Si NumPy es una calculadora científica (base matemática), Pandas es una hoja de cálculo inteligente que sabe el nombre de cada columna y puede manejar texto.


Si Numpy es el motor de un coche, Pandas es el volante, los asientos y el GPS.

## **La Serie — una columna con índice**

Una `Series` es exactamente eso: **una columna de datos con una etiqueta (índice) para cada fila**.

```python
#Crear una serie desde una lista
edades = pd.Series([25 , 30 , 28])
print(edades)
```

```python
#Se puede personalizar el índice (las 'etiquetas' de cada fila)
edades = pd.Series([25 , 30 , 28])
index = ["Ana" , "Luis" , "Maria"]
edades.index = index
print(edades)    #Imprime los datos en formato de lista
edades           #Imprime los datos en forma de tabla
```

```python
#type
print('\nTipo de objeto:', type(edades))
```

```python
# Acceder a un valor por índice de posición
print(edades[1])
```

```python
# Acceder a un valor por etiqueta
print(edades["Luis"])
```

## Creación del objeto Series

```python
# Creacion de un objeto Series
s = pd.Series([2, 4, 6, 8, 10])
print(s)
```

```python
# Creación de un objeto Series inicializándolo con un diccionario de Python
altura = {"Santiago" : 170 , "Pedro" : 178 , "Julia" : 187 , "Ana" : 165}
s = pd.Series(altura)
print(s)
```

```python
# Creación de un objeto Series inicializándolo con algunos
# de los elementos de un diccionario de Python
altura = {"Santiago" : 170 , "Pedro" : 178 , "Julia" : 187 , "Ana" : 165}
s = pd.Series(altura , index = ["Julia" , "Ana"])
print(s)
```

## Acceso a los elementos de un objeto Series

Cada elemento en un objeto Series tiene un identificador único que se denomina **_index label_**.

```python
# Creación de un objeto Series con index
s = pd.Series([2, 4, 6, 8, 10], index=["a", "b", "c", "d", "e"])
print(s)
```

```python
# Accediendo al tercer elemento del objeto
s ["c"]
```

```python
# Tambien se puede acceder al elemento por posición
s [2]
```

```python
# loc es la forma estándar de acceder a un elemento de un objeto Series por atributo
s["e"]
```

```python
# iloc es la forma estándar de acceder a un elemento de un objeto Series por posición
s.iloc[2]
```

```python
# Accediendo al segundo y tercer elemento por posición
s.iloc[2 : 4]
```

## Operaciones aritméticas con Series

```python
# Creacion de un objeto Series
s = pd.Series([2, 4, 6, 8, 10])
print(s)
```

```python
# Los objeto Series son similares y compatibles con los Arrays de Numpy
import numpy as np
# Ufunc de Numpy para sumar los elementos de un Array
np.sum(s)
```

```python
# El resto de operaciones aritméticas de Numpy sobre Arrays también son posibles
# Más información al respecto en la Introducción a Numpy
s * 2
```

## Representación gráfica de un objeto Series

## Creación de un objeto DataFrame: El DataFrame — la tabla completa

Un `DataFrame` es una **tabla de datos** con filas y columnas nombradas.  
Imagínalo como un diccionario de Series, donde cada clave es el nombre de una columna.



image.png

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])  #Sale NaN porque no aporta dato para el análisis o
                                                    #no encuentra el dato por formato
}

df = pd.DataFrame(personas)
df
```

```python
#Podemos ver las primeras 2 filas
print(df.head(2))
```

```python
# últimas filas
print(df.tail(2))
```

```python
#Ver dimensiones y estructura básica
# . shape - Devuelve una tupla (filas.columnas)
print("Filas y columnas: ", df.shape)

print ("\n", "-------------------" )
#Nos devuelve tipos de datos de cada columna. Sirve para verificar que el formato de las columnas es el correcto
print(df.dtypes)
```

Puede forzarse al DataFrame a que presente unas columnas determinadas y en un orden determinado

```python
# Creación de un DataFrame inicializándolo con algunos elementos de un diccionario
# de objetos Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(    #Filtros de datos que van a aparecer
        personas,
        columns = ["altura", "peso"],
        index = ["Ana", "Julia", "Santiago"])
```

## Copiar un DataFrame

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
# Copia del DataFrame df en df_copy
# Importante: Al modificar un elemento de df_copy no se modifica df
df_copy = df.copy()
df_copy
```

## Modificación de un DataFrame

```python
df_copy
```

```python
# Añadir una nueva columna calculada al DataFrame
df_copy["años"] = 2026 - df_copy["cumpleaños"]
df_copy
```

```python
# Eliminar una columna existente del DataFrame
del df_copy["peso"]
df_copy
```

```python
# Eliminar una columna existente devolviendo una copia del DataFrame resultante
df_mod = df.drop(["hijos"], axis = 1) #Toma en cuenta el df original
                                      #axis=1 opera en columnas y axis=0 opera en filas
df_mod
```
