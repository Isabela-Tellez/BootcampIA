### Respuesta 2 - Exploración de Herramientas (Postman)

Exploración de Herramientas (Postman)
Headers típicos en la respuesta:

```python
"""HTTP/1.1 200 OK
Content-Type: application/json; charset=utf-8
Cache-Control: public, max-age=86400
Date: Tue, 14 Apr 2026 12:00:00 GMT"""
```

¿Qué significa cada uno?
- Content-Type: Indica que la respuesta es un JSON.
- Cache-Control: El servidor sugiere que esta respuesta puede guardarse en caché durante 24 horas (86400 segundos).
- Date: La fecha y hora en que se generó la respuesta.

### Respuesta 3 - El Protocolo (HTTP/1.1)

¿Por qué HTTP/1.1 y no HTTP/2 o HTTP/3?
- Compatibilidad: HTTP/1.1 es universal. Todos los servidores, proxies y navegadores lo entienden.
- Simplicidad: Para APIs RESTful, HTTP/1.1 es suficiente. No necesitas la multiplexación de HTTP/2 si solo envías peticiones simples.
- Inercia: Cambiar de versión requiere actualizar servidores y clientes. Muchas APIs prefieren la estabilidad.

### Respuesta 7 - Criterio Ético: Rate Limiting

¿Qué pasaría con 1.000.000 de peticiones en un segundo?
- Sobrecarga del servidor: El servidor se saturaría y dejaría de responder a otros usuarios.
- Bloqueo de IP: La API detectaría el ataque y bloquearía tu IP temporal o permanentemente.
- Coste económico: Para el dueño de la API, cada petición consume recursos (CPU, ancho de banda).

¿Qué es el Rate Limiting?
- Es un mecanismo que limita el número de peticiones que un cliente puede hacer en un período de tiempo (ej: 100 peticiones por minuto).
- Ejemplo de header de Rate Limiting:

```python
cabeceras_rate_limit = """
X-RateLimit-Limit: 100
X-RateLimit-Remaining: 95
X-RateLimit-Reset: 30
"""
```

  - Limit: Máximo de peticiones permitidas.
  - Remaining: Peticiones restantes antes de ser bloqueado.
  - Reset: Segundos hasta que se reinicie el contador.

### Explicación detallada de cada parte:


Cabeceras (Headers):
1. HTTP/1.1 200 OK:
   - 200 OK: Código de estado que indica que la petición fue exitosa.
   - HTTP/1.1: Versión del protocolo HTTP utilizada.

2. Content-Type: application/json; charset=utf-8:
   - Indica que el cuerpo de la respuesta está en formato JSON y usa codificación UTF-8.

3. Content-Length: 647:
   - Tamaño del cuerpo de la respuesta en bytes (647 bytes en este caso).

4. Connection: close:
   - Indica que la conexión se cerrará después de esta respuesta (no se mantendrá abierta para futuras peticiones).

5. Server: netlify:
   - Informa que el servidor que está sirviendo la respuesta es Netlify (plataforma de hosting).

6. Access-Control-Allow-Origin: *:
   - Permite que cualquier dominio (representado por *) acceda a este recurso. Es una cabecera de CORS (Cross-Origin Resource Sharing) que evita problemas de acceso desde diferentes dominios.

### Cuerpo (Body):


El cuerpo de la respuesta es un JSON con los datos del personaje Rick Sánchez. Aquí está desglosado:

```python
{
  "id": 1,
  "name": "Rick Sanchez",
  "status": "Alive",
  "species": "Human",
  "type": "",
  "gender": "Male",
  "origin": {
    "name": "Earth (C-137)",
    "url": "https://rickandmortyapi.com/api/location/1"
  },
  "location": {
    "name": "Citadel of Ricks",
    "url": "https://rickandmortyapi.com/api/location/3"
  },
  "image": "https://rickandmortyapi.com/api/character/avatar/1.jpeg",
  "episode": [
    "https://rickandmortyapi.com/api/episode/1",
    "https://rickandmortyapi.com/api/episode/2"
  ],
  "url": "https://rickandmortyapi.com/api/character/1",
  "created": "2017-11-04T18:48:46.250Z"
}
```

Campos destacados:
- id: Identificador único del personaje (1 para Rick Sánchez).
- name: Nombre del personaje.
- origin y location: Objetos que contienen información sobre el origen y la ubicación actual del personaje.
- episode: Lista de URLs que apuntan a los episodios en los que aparece el personaje.
- created: Fecha y hora en que se creó este registro en la base de datos.

### Extracción del Nombre en Python
Si se quisiere extraer el nombre de Rick Sánchez del JSON en Python, se puede hacerlo de esta manera:

```python
import json

# Respuesta simulada (cuerpo de la respuesta)
respuesta_json = '''
{
  "id": 1,
  "name": "Rick Sanchez",
  "status": "Alive",
  "species": "Human",
  "type": "",
  "gender": "Male",
  "origin": {
    "name": "Earth (C-137)",
    "url": "https://rickandmortyapi.com/api/location/1"
  },
  "location": {
    "name": "Citadel of Ricks",
    "url": "https://rickandmortyapi.com/api/location/3"
  },
  "image": "https://rickandmortyapi.com/api/character/avatar/1.jpeg",
  "episode": [
    "https://rickandmortyapi.com/api/episode/1",
    "https://rickandmortyapi.com/api/episode/2"
  ],
  "url": "https://rickandmortyapi.com/api/character/1",
  "created": "2017-11-04T18:48:46.250Z"
}
'''

# Convertir el JSON a un diccionario de Python
datos = json.loads(respuesta_json)

# Extraer el nombre
nombre = datos["name"]
print(nombre)  # Salida: "Rick Sanchez"
```

- Depuración: Saber leer las cabeceras y ayuda a identificar problemas (ej: si Content-Type no es application/json, el código podría fallar al parsear la respuesta).
- Optimización: Content-Length permite estimar el tamaño de los datos recibidos.
- Seguridad: Access-Control-Allow-Origin es clave para entender cómo se gestionan los permisos de acceso a la API.
