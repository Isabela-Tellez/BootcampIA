## BLOQUE 1 · Tendencia Central: Media, Mediana y Moda

**Medidas de Tendencia Central** (El Corazón de los Datos)

"Existen tres tipos:

- **La Media** (Promedio): Es la democracia. Sumamos todo el dinero y lo repartimos a partes iguales entre todos. Es justa, pero muy sensible. Si entra un millonario, la media se dispara.

- **La Mediana**: Es el hermano del medio. Si ordenamos a todos los empleados de menor a mayor sueldo, la mediana es el que está justo en el centro. A la mediana no le importa si el último gana millones. Es resistente a los rebeldes.

- **La Moda**: Es el Influencer. Es simplemente el valor que más se repite. Lo que está de moda."

## ⚠️ El problema de la media: los valores extremos (outliers)

La media tiene una **debilidad importante**: un solo valor muy alto o muy bajo la **distorsiona** completamente.  
Esto es crítico en Ciencia de Datos — una media engañosa puede llevarnos a conclusiones incorrectas.

---
### 📐 La MEDIANA

**Idea:** Ordena todos los valores de menor a mayor. La mediana es el del **medio**.

Si el número de datos es PAR, es la media de los dos del centro.

```python
np.median(array)
```

**Intuición:** Si hiciéramos una fila ordenada de menor a mayor sueldo, la mediana es el sueldo de la persona que está justo en el centro de la fila.

| Tipo de datos | Qué pasa             | Fórmula           | Resultado     |
| ------------- | -------------------- | ----------------- | ------------- |
| Impar         | Hay un centro exacto | `(n + 1)/2`       | Un solo valor |
| Par           | Hay dos centros      | `n/2` y `n/2 + 1` | Promedio      |

````python
print(f"Salarios: {salarios}")

#mediana
np.median(salarios)
print (f"Mediana: {np.median(salarios)}")
````

````python
#Ordenar valores menor a mayor
salarios_ordenados = np.sort(salarios)
print (f"Salarios ordenados: {salarios_ordenados}")
````

## BLOQUE 2 · Dispersión: Desviación Estándar

###  La Desviación Estándar

**¿Qué es?** Mide cuánto se *dispersan* los datos respecto a la media. Una desviación alta = notas muy distintas entre sí. Una desviación baja = notas muy similares.

$$\sigma = \sqrt{\frac{\sum(x_i - \bar{x})^2}{n}}$$

>  **Piénsalo así:** dos clases pueden tener la misma media de 6.0, pero en una todos sacaron entre 5.5 y 6.5, y en la otra entre 2 y 10. La desviación estándar te cuenta esa historia.

**Analogía cotidiana:**

Dos pizzerías tienen el mismo tiempo de entrega promedio: **30 minutos**.

¿Son iguales de buenas? Vamos a ver...

**Pizzería A**: Tarda siempre 30 minutos. (Perfecto).

**Pizzería B**: A veces tarda 10 minutos, a veces 50 minutos. El promedio también es 30... ¡pero es un caos!

Para medir el 'caos' o la 'consistencia', usamos la Varianza y la Desviación Estándar."

````python
#Datos
pizzeria_A = np.array([29, 30, 31, 30, 29, 31]) #Son valores muy seguidos
pizzeria_B = np.array([10, 50, 5, 55, 30, 30]) #Son valores muy dispersos

#Media
print(f"\nMedia pizzeria_A: {np.mean(pizzeria_A)}")
print(f"Media pizzeria_B: {np.mean(pizzeria_B)}")

#Mediana
np.median(pizzeria_A)
print (f"\nMediana pizzeria_A: {np.median(pizzeria_A)}")
np.median(pizzeria_B)
print (f"Mediana pizzeria_A: {np.median(pizzeria_B)}")

#Moda
resultado_A = stats.mode(pizzeria_A)
resultado_B = stats.mode(pizzeria_B)
print (f"\nModa pizzeria_A: {resultado_A.mode}")
print (f"Moda pizzeria_B: {resultado_B.mode}")

#Desviación estándar
print (f"\nDesviación estándar pizzeria_A: {np.std(pizzeria_A): .3f}(Muy predecible)")
print (f"Desviación estándar pizzeria_B: {np.std(pizzeria_B): .3f}(Una lotería)")
````

````python
print(f"Variance Pizzeria A: {np.var(pizzeria_A):.2f}")

print(f"Variance Pizzeria B: {np.var(pizzeria_B):.2f}")
````

## 📋 Tabla resumen — ¿Cuándo usar qué?

| Medida | Pregunta que responde | Úsala cuando... | Evítala cuando... |
|--------|----------------------|-----------------|-------------------|
| **Media** | ¿Cuál es el valor típico? | Los datos son simétricos y sin outliers | Hay valores extremos (sueldos, precios) |
| **Mediana** | ¿Dónde está el centro real? | Hay outliers o datos sesgados | Los datos son perfectamente simétricos |
| **Moda** | ¿Qué ocurre más? | Datos categóricos o quieres saber lo más frecuente | Los datos son continuos y muy variados |
| **Desviación estándar** | ¿Cuánto se dispersan? | Quieres medir riesgo, homogeneidad, calidad | Se usa siempre junto a la media |
