## Conceptos clave antes de empezar


### ¿Qué es el Web Scraping?
Es la técnica de **extraer datos de páginas web automáticamente** con código. En lugar de copiar y pegar a mano, escribimos un programa que lo hace por nosotros.

### ¿Qué herramientas vamos a usar?

| Librería | ¿Para qué sirve? |
|---|---|
| `requests` | Para **pedir** la página web (como cuando tú abres una URL en el navegador) |
| `BeautifulSoup` | Para **leer y buscar** dentro del HTML de esa página |

### ¿Qué es el HTML?
Es el lenguaje con el que están construidas las páginas web. Tiene **etiquetas** como `<h3>`, `<p>`, `<div>`. BeautifulSoup nos ayuda a navegar por esas etiquetas.

## 🌐 SECCIÓN 2 — Hacer la petición a la web

🕵️ **Estableciendo el "Apretón de Manos" con el Servidor**

Antes de poder extraer datos, se debe "llamar a la puerta" de la web y esperar a que abra. En el mundo del desarrollo, esto se conoce como realizar una petición HTTP.

La misión es completar las líneas faltantes para conectar el script con el catálogo de libros. El objetivo es recibir un Código 200 (OK).

📋 Instrucciones:
 - Define el objetivo: Asigna la dirección http://books.toscrape.com/ a la variable de texto.

 - Solicita el acceso: Utiliza la librería requests para enviar una petición de tipo GET a esa URL.

 - Verifica la respuesta: Ejecuta el código para comprobar si el servidor te da permiso para entrar.

```python
# Guardad la URL en una variable llamada 'url'
url = "http://books.toscrape.com/"

# Investigad requests.get() para usarla, hacer la petición y guardadla en 'respuesta'
respuesta = requests.get(url)

# Esta línea comprueba que todo fue bien (200 = OK)
print("Código de respuesta:", respuesta.status_code)

# Si se ve 200, ¡está conectados! Si se ve otro número, algo salió mal.
if respuesta.status_code == 200:
    print("✅ ¡Conexión exitosa!")
else:
    print("❌ Algo salió mal. Revisad la URL.")
```

## 🍲 SECCIÓN 3 — Parsear el HTML con BeautifulSoup

Si tiene el contenido HTML de la página en `respuesta.text`. Ahora se le tiene que pasar a **BeautifulSoup** para que lo organice y se pueda buscar fácilmente.

**Paso 4.** Crear el objeto `soup`.

```python
# Cread el objeto BeautifulSoup
# PISTAS: Investigad y entented lo siguiente :
#             respuesta.text como contenido
#             "html.parser" como parser
soup = BeautifulSoup(respuesta.text, "html.parser")

print("✅ HTML parseado correctamente")
print("Tipo del objeto soup:", type(soup))
```

**Paso 5.** Antes de buscar datos, inspeccionar el HTML para entender su estructura.

> 💡 Se abre [books.toscrape.com](http://books.toscrape.com) en el navegador, luego clic derecho sobre un libro → **Inspeccionar**. Observar qué etiquetas envuelven el título, las estrellas y el stock.

```python
# Cómo visualizar los primeros 5000 caracteres del HTML generados en el paso anterior
# Eso ayudará a entender la estructura antes de buscar datos
# Código aquí:
print(soup.prettify()[:5000])
```

Ese texto con etiquetas como `<html>`, `<head>`, `<body>` es **HTML**: el lenguaje con el que están escritas todas las páginas web. Para nosotros es difícil de leer así, pero BeautifulSoup lo organiza y nos permite buscar dentro de él.

## 📦 SECCIÓN 4 — Extraer los datos

Ahora viene la parte central del reto: **extraer los datos** de cada libro.

Cada libro en el HTML está dentro de una etiqueta `<article class="product_pod">`. Vamos a buscar todos los artículos y luego extraer los datos de cada uno.

**Paso 6.** Buscad todos los libros de la página.

```python
# Investigar y usar soup.find_all() para encontrar todos los elementos de la primera página
# PISTA: con la etiqueta 'article' y la clase 'product_pod'
libros = soup.find_all("article", class_="product_pod")

print(f"✅ Encontrados {len(libros)} libros en la página")
```

**Paso 7.** Antes de hacer el bucle completo, trabajad con **un solo libro** para entender la estructura.

**La Lupa (Inspección del Contenedor)**


No podemos extraer datos a ciegas. En este paso, vamos a aislar el primer elemento de nuestra lista y usar el método **.prettify().** Este método organiza el código HTML de forma indentada y limpia, permitiéndonos ver la estructura interna (etiquetas, clases y atributos) de un solo libro.

**📋 Instrucciones:**
Selecciona la muestra:
 - Accede al primer elemento (índice 0) de tu lista de libros.

 - Imprime la estructura: Usa .prettify() para visualizar el "esqueleto" de ese libro en la consola.

 - Misión de observación: Mientras se imprime el código, intenta localizar visualmente:

   - ¿En qué etiqueta está el título?

   - ¿Qué clase tiene el párrafo del precio?

   - ¿Cómo se indica el rating en las clases CSS?

```python
# Cogemor solo el primer libro para explorar
# Investigar cómo funciona y usar prettify e imprimir
primer_libro = libros[0]
print(primer_libro.prettify())
```

```python
# Extraed el TÍTULO del primer libro
# Pista: el título está en el atributo 'title' de una etiqueta <a> dentro de <h3>
#output esperada: "Título: A Light in the Attic"
titulo = primer_libro.find("h3").find("a")["title"] #find("h3") → localiza el contenedor del títuloç
                                                    #efind("a") → encuentra el enlace dentro
                                                    #["title"]  → extrae el atributo donde está el texto del título

print(f"Título: {titulo}")
```

```python
# Extraer el RATING del primer libro
# Pista: está en una etiqueta <p> con clase que empieza por 'star-rating'
# El rating está en el segundo elemento de las clases de esa etiqueta: .get('class')[1]
rating_raw = primer_libro.find("p", class_="star-rating").get("class")[1]  #find("p", class_="star-rating") → localiza el <p> del rating
                                                                           #.get("class") → devuelve una lista como ["star-rating", "Three"]
                                                                           #[1] → toma el segundo elemento → "Three"

print("Rating (texto):", rating_raw)

# Convertir el texto a número usando este diccionario
# Tener en cuenta la primera línea Mayúscula para que se reconozca
convertir = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
rating = convertir[rating_raw]
print("Rating (número):", rating)



# Output esperada:
#     Rating (texto): Three
#     Rating (número): 3
```

```python
# Extraer la DISPONIBILIDAD del primer libro
# Pista: está en una etiqueta <p> con clase 'instock availability'
# Investigad y usad .text.strip() para limpiar espacios
disponibilidad = primer_libro.find("p", class_="instock availability").text.strip()  #find("p", class_="instock availability") → encuentra la etiqueta correcta
                                                                                     #.text → obtiene el texto dentro (incluye saltos y espacios)
                                                                                     #.strip() → limpia espacios y saltos de línea


print("Disponibilidad:", disponibilidad)
```

**Paso 8.** ¡Ahora hacedlo para **todos los libros** con un bucle!

```python
# Diccionario para convertir rating de texto a número
convertir = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

# Lista donde guardaremos todos los resultados
catalogo = []

# Completad el bucle
for libro in libros:
    titulo = libro.find("h3").find("a")["title"]
    rating = convertir[libro.find("p", class_="star-rating").get("class")[1]]
    disponibilidad = libro.find("p", class_="instock availability").text.strip()

    # Guardamos los datos como un diccionario
    catalogo.append({
        'titulo': titulo,
        'rating': rating,
        'disponibilidad': disponibilidad
    })

print(f"✅ Extracción completada: {len(catalogo)} libros procesados")
print("\n📚 Primeros 3 libros:")
for libro in catalogo[:3]:
    print(libro)
```

## 🚀 Reto Extra

La página tiene **50 páginas** de libros. La URL de la segunda página es:
`http://books.toscrape.com/catalogue/page-2.html`

**¿Podéis modificar el código para extraer los libros de las primeras 3 páginas?**

Pistas:
- Investigad Paginación en Web scraping:
   - Usad un bucle `for` que itere sobre los números de página
   - Construid la URL dinámicamente con f-strings
   - Concatenad los resultados en una sola lista

```python
# 🚀 RETO EXTRA: Scraping de múltiples páginas
# Extraed libros de las páginas 1, 2 y 3
catalogo_completo = []
convertir = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

for pagina in range(1, 4):  # páginas 1, 2, 3
    if pagina == 1:
        url_pagina = "http://books.toscrape.com"
    else:
        url_pagina = f"http://books.toscrape.com/catalogue/page-{pagina}.html"  # TODO: construid la URL

    # TODO: repetid el proceso de scraping para cada página
    respuesta = requests.get(url_pagina)
    soup = BeautifulSoup(respuesta.text, "html.parser")

    libros = soup.find_all("article", class_="product_pod")

    for libro in libros:
        titulo = libro.find("h3").find("a")["title"]
        rating = convertir[libro.find("p", class_="star-rating").get("class")[1]]
        disponibilidad = libro.find("p", class_="instock availability").text.strip()

        catalogo_completo.append({
            "titulo": titulo,
            "rating": rating,
            "disponibilidad": disponibilidad
        })

print(f"Total de libros extraídos: {len(catalogo_completo)}")
```
